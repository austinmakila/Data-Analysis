```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent', 
                        missing_values=np.nan)
imputer = imputer.fit(data)
data.iloc[:,:] = imputer.transform(data)
data
```

```python
data['pre_usage']=data['pre_usage'].replace(['1 2','3 1','1 3','3'],[3,3,3,3])
data.pre_usage.value_counts()
```

```python
data['Primary_line']=data['Primary_line'].replace(['airtel','Airtel/Safaricom'],['Airtel','Airtel'])

data['Primary_line']=data['Primary_line'].replace(['Telcomm'],['Telkom'])
data['Primary_line']=data['Primary_line'].replace(['safcom'],['Safaricom'])
data['Primary_line']=data['Primary_line'].replace(['Telcom'],['Telkom'])





data.Primary_line.value_counts()
```

```python
data['lte_type']=data['lte_type'].replace(['2 3','1 2','1 2 3','2 1','3 2','4 3','2 3 1','3 4','4','2','3'],[3,3,3,2,3,4,3,4,4,2,3])
data.lte_type.value_counts()
```

```python
data['pre_usage']=data['pre_usage'].replace(['1 2','1 3','3 1','3','2'],[2,3,3,3,2])
data.pre_usage.value_counts()
```

```python
data['Monthly_income']=data['Monthly_income'].replace(['10_000___30_000','__10_000','30_001___50_000','__50_000'],
                                                  ['10000-30000','<10000','30001-50000','>50000'])
data.Monthly_income.value_counts()
```
